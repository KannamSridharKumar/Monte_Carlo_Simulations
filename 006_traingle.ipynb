<b> <font size="5" color="white">Line to Traingle</font> </b>

- Assume that we have a line of length 1 meter. Cut the line into 3 segements by selecting 2 points on the line. 

- What's the probability that the resulting three lines/segments can form a triangle?



```python
%load_ext nb_black
import IPython
import random
```

```python
path = "images/line_2_traingle.jpeg"
IPython.display.Image(path, width=600)
```

```python
n_triangles = 0
n_runs = 1000000
```

```python
for _ in range(n_runs):
    x1, x2 = random.random(), random.random()

    if x2 > x1:
        s1, s2, s3 = x1, x2 - x1, 1 - x2
    else:
        s1, s2, s3 = x2, x1 - x2, 1 - x1

    if s1 + s2 > s3 and s1 + s3 > s2 and s2 + s3 > s1:
        n_triangles += 1
```

```python
print("Total number of trials: ", n_runs)
print("Number of triangles: ", n_triangles)
print("probability of a triangle: ", n_triangles / n_runs)
```
