```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def random_walk(n_steps):
    xl, yl = [0], [0]
    x, y = 0, 0
    for i in range(n_steps):
        step = random.choice(["N", "E", "S", "W"])
        # step = random.choice(['Up', 'Right', 'Down', 'Left'])
        if step == "N":
            y = y + 1
        elif step == "S":
            y = y - 1
        elif step == "E":
            x = x + 1
        else:
            x = x - 1

        xl.append(x)
        yl.append(y)
    return xl, yl
```

```python
def random_walk_2(n_steps):
    xl, yl = [0], [0]
    x, y = 0, 0
    for i in range(n_steps):
        dx, dy = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy

        xl.append(x)
        yl.append(y)
    return xl, yl
```

```python
n_steps = 100
xl, yl = random_walk_2(n_steps)
```

```python
print("final location: ", xl[-1], yl[-1])

for i in range(len(xl) - 1):
    plt.plot([xl[i], xl[i + 1]], [yl[i], yl[i + 1]], "-", lw=2, color="red")

bound = 10
plt.xlim(-bound, bound)
plt.ylim(-bound, bound)
plt.xticks(range(-bound, bound))
plt.yticks(range(-bound, bound))
plt.grid()
plt.tight_layout()
```

```python
xl2 = []
yl2 = []
fig, ax = plt.subplots()


def animate(i):
    xl2.append(xl[i])
    yl2.append(yl[i])
    ax.clear()
    ax.plot(xl2, yl2, "-", lw=2, color="red")
    bound = 40
    ax.set_title(
        "step: " + str(i + 1) + ", location: (" + str(xl[i]) + ", " + str(yl[i]) + ")"
    )
    ax.set_xlim(-bound, bound)
    ax.set_ylim(-bound, bound)
    ax.set_xticks(range(-bound, bound))
    ax.set_yticks(range(-bound, bound))
    ax.grid()
```

```python
ani = FuncAnimation(fig, animate, frames=n_steps, interval=10, repeat=False)
plt.show()
```
