```python
doors = [1, 2, 3]
```

```python
car_door = random.choice(doors)
print("car door: ", car_door)
goat_doors = list(set(doors) - set([car_door]))
print("goat doors: ", goat_doors)
```

```python
def monte_hall(n_runs, switch):
    n_wins = 0

    for _ in range(n_runs):
        initial_selection = random.choice(doors)

        host_options = list(set(goat_doors) - set([initial_selection]))
        host_selection = random.choice(host_options)

        if switch == "no":
            final_selection = initial_selection

        if switch == "yes":
            final_selection = list(set(doors) - set([initial_selection, host_selection]))[0]

        #print(initial_selection, host_selection, final_selection)
        #print("initial: {}, host: {}, final: {}".format(initial_selection, host_selection, final_selection))

        if car_door == final_selection:
            n_wins += 1


    #print(n_wins / n_runs)
    return n_wins/n_runs
```

```python
n_runs = 100000
switch = "no"
print("-----Without Switching-----")
prob = monte_hall(n_runs, switch)
print("\n Winning Prob: ", prob)
print("\n\n")
print("-----With Switching-----")
switch = "yes"
prob = monte_hall(n_runs, switch)
print("\n Winning Prob: ", prob)
```
