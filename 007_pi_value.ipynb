```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
n_drops = 100000
x = np.random.uniform(low=-1, high=1, size=[n_drops,1])
y = np.random.uniform(low=-1, high=1, size=[n_drops,1])
print(x[:10])
```

```python
is_inside = x**2 + y**2 < 1
print(is_inside[:10])
```

```python
n_inside = np.sum(is_inside)
print(n_inside)
```

```python
prob_inside = n_inside / n_drops

print("Total number of drops: ", n_drops)
print("Number of drops inside the circle: ", n_inside)
print("probability of dropping inside the circle: ", prob_inside)
```

```python
circle_radius = 1.0
square_side = 2.0
circle_area = np.pi * circle_radius * circle_radius
square_area = square_side * square_side
```

- prob_inside = circle_area / square_area = pi * 1 * 1 / (2 * 2)
- prob_inside = pi / 4
- pi = 4 * prob_inside

```python
pi = 4*prob_inside
print(np.pi, pi)
```

```python
x_in = x[is_inside]
y_in = y[is_inside]
plt.figure(figsize=[7,7])
plt.scatter(x, y, s=1)
plt.scatter(x_in, y_in, color='g', s=1)
```
