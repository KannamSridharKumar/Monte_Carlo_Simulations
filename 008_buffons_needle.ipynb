<b> <font size="5" color="white">Buffon's Needle</font> </b>




- Find the probability that a needle intersects a line. 
- An estimate of PI value    

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
plt.style.use("fivethirtyeight")
```

```python
n_runs = 1000000
n_hits = 0
```

```python
xi = []
xj = []
yi = []
yj = []
```

```python
for i in range(n_runs):
    
    x1 = 9 * np.random.rand() + 1
    y1 = 9 * np.random.rand() + 1

    theta = 2*np.pi * np.random.rand()

    x2 = x1 + np.cos(theta)
    y2 = y1 + np.sin(theta)

    check = np.ceil(np.minimum(y1, y2)) == np.floor(np.maximum(y1, y2))
    n_hits += check
    
    xi.append(x1)
    xj.append(x2)
    yi.append(y1)
    yj.append(y2)
```

```python
prob = n_hits / n_runs
print('Total number of needles/runs: ', n_runs)
print('Number of needles intersected: ', n_hits)
print('Probability of intersection: ', prob)
print('pi: ', 2 / prob)
```

```python
plt.figure(figsize=[7,7])
plt.plot([xi, xj], [yi, yj], '-')
plt.axis([0, 11, 0, 11])
plt.xticks(range(12))
plt.yticks(range(12))
plt.gca().yaxis.grid(True, linewidth=3, color="blue")
```
